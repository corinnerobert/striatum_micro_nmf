```python
import sys, os
import glob
import pandas as pd
import numpy as np
import scipy, scipy.io


%pylab inline
```

```python
import scipy, scipy.io
pls_dir = ""
fname = pls_dir + "left_lv_one.mat" #load results for an lv

pls_res = scipy.io.loadmat(fname)
#Build a list of behavs and components text once, will be same for all
#this is used in plotting for axis labels
components_list = []
for idx in range(0,len(pls_res['components'].tolist()[0])):
    components_list.append( str(pls_res['components'].tolist()[0][idx][0]) )

    behavs_list = []
for idx in range(0,len(pls_res['behavs_text'].tolist()[0])):
    behavs_list.append( str(pls_res['behavs_text'].tolist()[0][idx][0]) )
    
```

```python
#load lv one results
fname = pls_dir + "left_lv_one.mat"
left_lv_one = scipy.io.loadmat(fname)

#get the lower and upper bounds of the conf intervals for behav variables
err = np.concatenate( (left_lv_one['lower'], left_lv_one['upper']))
```

```python
#Create brain and behav plots for lv one.  This chunk of text should create two plots
#on left, behavrious variables - for each variable the point estimate of the correlations isplotted 
#as well as the 95% conf interval

#for brain variables, the BSR for each brain variable is plotted
#a black line is also plotted at x = 2.58 and x=-2.58
#this value corresponds a 99% conf intervals for brain variables
# if BSR crosses thi line -> brain variable contributes at that level (ie is 'significant')
#a threshold set at 1.96 is 95% conf interval. this is fine to work with as well

fig, ax = plt.subplots(2,1, figsize = (20,34),dpi = 300)
fig.suptitle('Left LV 1', y=0.95,fontsize=50)
plt.subplots_adjust(wspace = 0.3)

#behaviour plot
y_pos = np.arange(len(left_lv_one['lvcorrs'].tolist()[0]))
#make bar plot - y axis has each behav variable, x axis is the correlation (lvcorrs variable)
ax[0].barh(y_pos,left_lv_one['lvcorrs'].tolist()[0], xerr = err, align = 'center')
#set y labels to the behavs_list
ax[0].yaxis.set_major_formatter(FixedFormatter(behavs_list)) 
ax[0].yaxis.set_major_locator(FixedLocator(y_pos))
ax[0].tick_params(axis='both', labelsize=50)
ax[0].grid(which='both', color='#CCCCCC')
ax[0].set_facecolor('white')


#brain plot
y_pos = np.arange(len(left_lv_one['bsr'].tolist()[0]))
#make bar pot - y axis is each brain variable, x axis is BSR
colors = ['#9B81C7','#9B81C7','#9B81C7','#8F0748','#8F0748','#8F0748','#75CCAA','#75CCAA','#75CCAA',
          '#EA8B0D','#EA8B0D','#EA8B0D','#013B25','#013B25','#013B25']
ax[1].barh(y_pos,left_lv_one['bsr'].tolist()[0], align = 'center',color=colors)
ax[1].yaxis.set_major_formatter(FixedFormatter(components_list))
ax[1].yaxis.set_major_locator(FixedLocator(y_pos))
ax[1].tick_params(axis='both', labelsize=45)
#the if and axvline cmds plot the black line at 2.58
if( np.max(left_lv_one['bsr']) > 2.5):
    ax[1].axvline(x=2.58, color = 'black', ymin=0.03, ymax = 0.97)
if( np.min(left_lv_one['bsr']) < -2.5):
    ax[1].axvline(x=-2.58, color = 'black', ymin=0.03, ymax = 0.97)

ax[1].grid(which='both', color='#CCCCCC')
ax[1].set_facecolor('white')
plt.savefig(pls_dir + 'left_lv_one_big_v2.png', bbox_inches='tight') #save to png, modify file name if you like

```
